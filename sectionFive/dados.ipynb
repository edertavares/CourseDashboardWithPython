```python
import pandas as pd

data = {
    'Product': ['Camiseta', 'Calça', 'Tênis'],
    'Quantidade': [10, 15, 8],
    'Preço_Unitário': [29.99, 49.99, 89.99]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['Total'] = df['Quantidade'] * df['Preço_Unitário']
df
```

```python
df_pedidos = pd.read_csv("data/pedidos2.csv", sep=";")
df_pedidos 
```

```python
df_pedidos['Total'] = df_pedidos['Unidades'] * df_pedidos['PrecoUnidade']
df_pedidos
```

```python
vendas_por_regiao = df_pedidos.groupby('Regiao')['Total'].sum().reset_index()
vendas_por_regiao
```

```python
vendas_por_vendedor = df_pedidos.groupby('Vendedor')['Total'].sum().reset_index()
vendas_por_vendedor
```
